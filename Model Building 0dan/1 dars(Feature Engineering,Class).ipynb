# Feature Engineering

```python
import pandas as pd 
df=pd.read_csv('heart_disease_uci Classification.csv')
```

```python
df.info()
```

```python
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype=='object':
            df[col].fillna(df[col].mode()[0],inplace=True)
        else:
            df[col].fillna(df[col].mean(),inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

for col in df.columns:
    if df[col].dtype=='object':
        if df[col].nunique()<=5:
            dummies=pd.get_dummies(df[col],prefix=col,dtype=int)
            df=pd.concat([df.drop(columns=col),dummies],axis=1)
        else:
            df[col]=encoder.fit_transform(df[col])
```

```python
df.info()
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()

num_col=df.select_dtypes(include=['number']).columns.drop('num')
df[num_col]=scaler.fit_transform(df[num_col])
```

```python
df.head()
```

```python
# Feature Engineering

# 1. Qon bosimi va xolesterinni birlashtirib yurak xavfi ko'rsatkichi
df["risk_index"] = df["trestbps"] + df["chol"]

# 2. Yoshga nisbatan maksimal yurak urish tezligi
df["relative_hr"] = df["thalch"] / df["age"]

# 5. Metabolik xavf (qonda shakar + xolesterin)
df["fbs_encoded"] = df["fbs"].astype("category").cat.codes
df["metabolic_risk"] = df["fbs_encoded"] + df["chol"]
```

```python
df.info()
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
```

```python
class Qutbiddin:
    def __init__(self,ism,yosh,millat,til,qiziqishlar):
        self.ism=ism
        self.yosh=yosh
        self.millat=millat
        self.til=til
        self.qiziqishlar=qiziqishlar

    def tanishtir(self):
        return f"Salom! Mening ismim {self.ism}.Men {self.yosh} yoshdaman,millatim {self.millat}."

    def tilni_ayt(self):
        return f"Men {self.til} tilida gaplashaman."

    def qiziqishlarni_ayt(self):
        return f"Mening qiziqishlarim: {', '.join(self.qiziqishlar)}."


# Ob'ekt yaratish
men=Qutbiddin("Qutbiddin",27,"o'zbek","o'zbek va koreys",["AI", "siyosat", "psixologiya", "din"])

# Methodlarni ishlatish
print(men.tanishtir())
print(men.tilni_ayt())
print(men.qiziqishlarni_ayt())
```
